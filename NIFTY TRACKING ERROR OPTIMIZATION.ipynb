```python
# Add Nifty index data to the main DataFrame
data["Nifty"] = Nifty.Nifty
```

```python
# Calculate returns
Returns = data.pct_change()
```

```python
# Function to calculate annual risk and return
def summary_ann(Return):
    summary = Return.agg(["mean", "std"]).T
    summary["Return"] = summary["mean"] * 252
    summary["Risk"] = summary["std"] * np.sqrt(252)
    summary.drop(columns=["mean", "std"], inplace=True)
    return summary
```

```python
# Calculate annual risk and return summary
summary = summary_ann(Returns)
summary.drop("Nifty", inplace=True)
```

```python
# Select a time period for tracking error analysis
data_2 = data.loc["2022":"2023"].copy()
Norm_data_2 = data_2.div(data_2.iloc[1]).mul(100)
Norm_data_2 = Norm_data_2.drop(Norm_data_2.index[0])
```

```python
# Calculate returns for the selected time period
Returns_2 = Returns.loc["2022":"2023"].copy()
```

```python
# Calculate return difference from the benchmark (Nifty index)
Return_diff = Returns_2.sub(Returns_2["Nifty"], axis=0)
```

```python
# Function to calculate premium and Tracking Error
def Tracking_ann(Return):
    summary = Return.agg(["mean", "std"]).T
    summary["Premium"] = summary["mean"] * 252
    summary["TE"] = summary["std"] * np.sqrt(252)
    summary.drop(columns=["mean", "std"], inplace=True)
    return summary
```

```python
# Calculate premium and Tracking Error
tracking = Tracking_ann(Return_diff)
tracking
```

```python
# Select a number of assets for portfolio construction
noa = 10
```

```python
# Select stocks with the least tracking error
tracking_stocks = tracking.sort_values(by="TE").index[:noa + 1]
tracking_stocks = tracking_stocks[1:]  # Exclude the Nifty index
```

```python
# Calculate equal weights
eweights = np.full(noa, 1 / noa)
```

```python
# Function to calculate portfolio return
def port_ret(weights):
    return Returns_2.loc[:, tracking_stocks].dot(weights)
```

```python

# Function to calculate tracking error
def tracking_error(weights):
    return port_ret(weights).sub(Returns_2.Nifty).std() * np.sqrt(252)
```

```python
# Calculate tracking error for equal-weighted portfolio
tracking_error(eweights)
```

```python
# Create equal-weighted portfolio
ew_port = port_ret(eweights).add(1).cumprod().mul(100)
norm_track = Norm_data_2.loc[:, tracking_stocks].copy()  # Make a copy to avoid modifying the original DataFrame
norm_track['Nifty'] = Norm_data_2['Nifty']  # Add the 'Nifty' column
norm_track["ew_port"] = ew_port
norm_track.iloc[0, -1] = 100
norm_track[['ew_port', 'Nifty']].plot(figsize=(15, 10))
plt.legend(fontsize=20, loc=2)
plt.show()
```

```python
# Portfolio optimization
cons = ({"type": "eq", "fun": lambda x: np.sum(x) - 1})
bounds = tuple((0, 1) for x in range(noa))
opts = sco.minimize(tracking_error, eweights, method="SLSQP", bounds=bounds, constraints=cons)
optimal_weights = opts["x"]
```

```python
# Create optimized portfolio
opt_port = port_ret(optimal_weights).add(1).cumprod().mul(100)
norm_track["opt_port"] = opt_port
norm_track.iloc[0, -1] = 100
norm_track[["opt_port", "Nifty"]].plot(figsize=(15, 10))
plt.legend(fontsize=20, loc=2)
plt.show()
```

```python
# Compare tracking error of equal-weighted and optimized portfolios
print("Tracking Error (Equal-Weighted):", tracking_error(eweights))
print("Tracking Error (Optimized):", tracking_error(optimal_weights))
```

```python
# Calculate risk and return for portfolios
summary_ann(norm_track[["opt_port", "ew_port", "Nifty"]].pct_change())
```

```python
# Function to calculate tracking error for a given time period
def tracking_error_gen(data, tracking_stocks, weights, index, start, end):
    return data.loc[start:end, tracking_stocks].dot(weights).sub(Returns.loc[start:end, index]).std() * np.sqrt(252)
```

```python
# Function to visualize portfolio performance
def visual(data, tracking_stocks, weights, index, start, end):
    track = data.loc[start:end, tracking_stocks].dot(weights).add(1).cumprod().mul(100)
    track.name = "Portfolio"
    index = data.loc[start:end, index].add(1).cumprod().mul(100)
    plt.figure(figsize=(12, 8))
    track.plot()
    index.plot()
    plt.legend(fontsize=20, loc=2)
    plt.show()
```

```python
# Calculate forward tracking error and visualize performance
forward_te = tracking_error_gen(Returns, tracking_stocks, optimal_weights, "Nifty", "2023-01-01", "2024-04-20")
print("Forward Tracking Error (2023-2024):", forward_te)
visual(Returns, tracking_stocks, optimal_weights, "Nifty", "2023-01-01", "2024-04-20")
```
